```python
import hicstraw
import numpy as np
import pandas as pd
```

```python
hic = hicstraw.HiCFile("./GSE89112_Kc167combined.hic")
print(hic.getChromosomes())
print(hic.getGenomeID())
print(hic.getResolutions())
```

```python
mzd = hic.getMatrixZoomData('chrX', 'chrX', "observed", "NONE", "BP", 500)
```

```python
two_classes = pd.read_csv("./dataset/fetaures_and_labels.2_classes.500bp_bin.tsv", sep = "\t")
X_only = two_classes[two_classes['chr']=='chrX']
X_only
```

```python
numpy_matrix = mzd.getRecordsAsMatrix(0, 2000000, 0, 2000000)
print(numpy_matrix.shape)
# Specify the filename and delimiter (optional, default is space)
filename = 'chrX_1000000_hic_4000.txt'
delimiter = ','  # For example, you can use a comma as a delimiter

# Save the matrix to a text file
np.savetxt(filename, numpy_matrix, delimiter=delimiter)
```

```python
train_epigenomic = X_only.iloc[0:4001,4:]
train_epigenomic.to_csv("./4000_epigemnomic_1000000.csv")
label_train =  X_only.iloc[0:4001,3].to_csv("./4000_label.csv")
```

```python
numpy_matrix = mzd.getRecordsAsMatrix(1000000, 2000000, 1000000, 2000000)
print(numpy_matrix.shape)
# Specify the filename and delimiter (optional, default is space)
filename = 'chrX_2000000_hic.txt'
delimiter = ','  # For example, you can use a comma as a delimiter

# Save the matrix to a text file
np.savetxt(filename, numpy_matrix, delimiter=delimiter)
```

```python
train_epigenomic = X_only.iloc[2000:4001,4:]
train_epigenomic.to_csv("./test_epigemnomic_2000000.csv")
label_train =  X_only.iloc[2000:4001,3].to_csv("./test_label.csv")
```
